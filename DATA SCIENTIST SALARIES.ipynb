```python
#Rename the Employment Type

ds['employment_type'] = ds['employment_type'].replace({'FT' :'Ful Time', 'CT' : 'Contract', 'PT': 'Part Time', 'FL':'Freelance'})

ds.head()
```

```python
# Rename the Experience Level

ds['experience_level'] = ds['experience_level'].replace({'MI' :'Mid-Level', 'SE' : 'Senior-Level', 'EN': 'Entry-Level', 'EX':'Executive-Level'})

ds.head()
```

```python
# Rename the Company size

ds['company_size'] = ds['company_size'].replace({'S' :'Small', 'L' : 'Large', 'M': 'Medium'})

ds.head()
```

###  Experience Level Analysis


3.1. Employment Type by Experience Level
3.2. Top 3 Job Title by Experience Level
3.3. Company Size by Experience Level


```python
ds.head()
```

```python
# Employment Type by Experience Level

sns.set_theme(font_scale=1.5)
plt.figure(figsize=(15,7))
plt.title("Employment Type by Experience Level")

sns.barplot( x ='employment_type', y = 'salary_in_usd', hue ='experience_level', data = ds)
```

```python
#How much each job title earn on the average

top5job = ds.groupby('job_title').salary_in_usd.mean().sort_values(ascending=False)[:5]
top5job
```

```python
sns.set_theme(font_scale=1.5)
top5job.plot.pie(autopct="%.1f%%",figsize=(7,7)).set_title("Top 5 Most Earned Job Titles")
sns.barplot( x ='employment_type', y = 'salary_in_usd', hue ='experience_level', data = ds)
```

```python
#Company Size by Experience Level

comp_size = ds.groupby('company_size').experience_level.count().sort_values(ascending=False)

comp_size
```

```python
#Top 3 Job Title by Experience Level

sns.set_theme(font_scale=1.5)
plt.figure(figsize=(14,5))
plt.title('Top 3 Job Title by Experience Level and Salary')

sns.barplot(x=(ds.job_title[:3]), y = 'salary_in_usd', hue ='experience_level', data =ds)
```

```python
#Company Size by Experience Level

sns.set_theme(font_scale=1.5)
plt.figure(figsize=(15,5))
plt.title("Company Size by Experience Level and Salary")

sns.barplot(x='company_size', y = 'salary_in_usd', hue ='experience_level', data =ds)
```
